```python
s=p.read_csv("student_performance.csv")
subject=["science","english","math"]
for sub in subject:
    d=s[sub]
    print("performance of students in",sub,":\nMean: ",n.mean(d),"\nMedian: ",n.median(d),"\nMode: ",n.argmax(n.bincount(d)),"\nStandard Deviation: ",n.std(d),"\n\n\n")
    m.boxplot(d)
    m.title("performance of students :")
    m.ylabel("scores")
    m.xlabel(sub)
    m.show()
```

```python
d=p.read_csv("sales_dataset.csv")
crr,pval=pearsonr(d["Sales"],d["ad"])
if crr>0:
    dir="positive"
elif crr<0:
    dir="Negative"
else:
    dir="no"
str=abs(crr)
print("sales revenue and advertising expenditure:\nCorrelation: ",crr,"\nP_Value: ",pval,"\nStrength: ",str,"\nDirection: ",dir)

m.scatter(d["Sales"],d["ad"])
m.title("\n\nsales revenue and advertising expenditure")
m.ylabel("Sales")
m.grid(True)
m.xlabel("ad")
m.show()
```

```python
d = {'sales': [2, 4, 6, 9, 12, 34, 45],'tv': [1, 2, 4, 7, 9, 11, 15]}
x=n.array(d['tv']).reshape(-1,1)
y=n.array(d['sales'])
model=LinearRegression()
model.fit(x,y)
print("\n\nCo-efficient",model.coef_,"\nIntercept",model.intercept_)
m.scatter(x,y,color="blue",label="Points")
m.plot(x,model.predict(x),color="red",label="Linear Regression")
m.title("\n\nsimple Linear Regression model")
m.ylabel("Sales")
m.xlabel("tv")
m.grid(True)
m.legend()
m.show()
```
